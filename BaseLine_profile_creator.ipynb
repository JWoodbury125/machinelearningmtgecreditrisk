```python
import pandas as pd
import numpy as np
import matplotlib as mlt
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
import statistics
```

```python
combined_state_data = pd.read_csv("State_Combined_Df.csv", header=0)
combined_state_data = combined_state_data.drop(['Unnamed: 0'], axis=1)
```

```python
for i in combined_state_data.index:
    combined_state_data.at[i,'AmorTerm'] = combined_state_data.at[i,'AmorTerm']
```

```python
combined_state_data['PMT'] = "Nan"
for i in combined_state_data.index:
    combined_state_data.at[i,'PMT'] = np.round(abs(np.pmt(rate=(combined_state_data.at[i,'Rate']/12), fv=0, nper=combined_state_data.at[i,'AmorTerm'], pv=combined_state_data.at[i,'Amount'])), 2)
```

```python
states = set(combined_state_data["StateName"])
clean_states = list(states)
clean_states.remove(clean_states[0])
```

```python
df = combined_state_data
df["PMT"] = df.PMT.astype(float)
```

```python
df1 = df[(df["Income"] > 10000) & (df1["Income"] < 400000)]
df1 = df1[(df1["IncRat"] != 1) & (df1["IncRat"] < 15)]
df = df1
```

```python
def createBaseLine(year):
    b_year = df[df["Year"] == year]
    b_year_temp = b_year.groupby("StateName").agg(lambda x: x.value_counts().index[0])
    byear = b_year_temp[["Income" , "IncRat", "Amount", "PMT"]]
    byear["Rate"] = b_year["Rate"].groupby(df["StateName"]).mean()
    byear["LTV"] = b_year["LTV"].groupby(df["StateName"]).mean()
    return byear
```

```python
baseline_2010 =  createBaseLine(2010)
baseline_2011 =  createBaseLine(2011)
baseline_2012 =  createBaseLine(2012)
baseline_2013 =  createBaseLine(2013)
baseline_2014 =  createBaseLine(2014)
baseline_2015 =  createBaseLine(2015)
baseline_2016 =  createBaseLine(2016)
baseline_2017 =  createBaseLine(2017)
```

```python
b = [baseline_2010,baseline_2011,baseline_2012,baseline_2013,baseline_2014,baseline_2015,baseline_2016,baseline_2017]
year = [2010,2011,2012,2013,2014,2015,2016,2017]
for base,year in zip(b,year):
    
    print("Baseline profile Perstate for year: ",year)
    print(base.head())
    print("\n\n")
```
