```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.display import display, HTML

%matplotlib inline

data = pd.read_csv("2017_PUDB_Export.csv")
```

#### Considering the important features

```python
interested = ['Bank', 'FIPSStateCode', 'FIPSCountyCode', 'MSA','Tract','MinPer', 'TraMedY', 'LocMedY', 'Tractrat',
       'Income', 'CurAreY', 'IncRat', 'UPB', 'LTV', 'Purpose','Coop','Product','FedGuar', 'Term', 'AmorTerm', 'SellType',
       'NumBor', 'First', 'CICA', 'BoRace', 'CoRace', 'BoGender', 'CoGender',
       'BoAge', 'CoAge', 'Occup', 'NumUnits', 'HOEPA', 'LienStatus', 'SpcHsgGoals', 'FedFinStbltyPlan', 'AcqTyp',
       'GSEREO','Geog', 'Amount',
       'Front', 'Back', 'BoCreditScor', 'CoBoCreditScor', 'PMI', 'Self',
       'PropType','BoEth', 'CoEth', 'Rate']
```

```python
new = data[interested]
```

```python
display(HTML(new.head().to_html()))
```

#### One Hot Encoding the categorical features.

```python
cat= ['Bank', 'FIPSStateCode','Purpose','Coop','Product','FedGuar', 'SellType',
       'NumBor', 'First', 'CICA', 'BoRace', 'CoRace', 'BoGender', 'CoGender',
       'Occup', 'NumUnits', 'HOEPA', 'LienStatus', 'SpcHsgGoals', 'FedFinStbltyPlan', 'AcqTyp',
       'GSEREO','Geog','Self', 'PropType']
new2 = pd.get_dummies(new, columns=cat, drop_first=True)
```

```python
new2.shape
```

```python
new2.columns
```

#### no_cat == Non Categorical Feature which we need to normalize

```python
no_cat = [s for s in interested if s not in cat]
```

```python
no_cat.remove("Rate")
```

#### Normalizing the Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scale = pd.DataFrame(data = new2)
scale[no_cat] = scaler.fit_transform(new2[no_cat])

# Show an example of a record with scaling applied
display(scale.head())
```

```python
rate = scale["Rate"]
```

```python
del scale["Rate"]
```

#### Train Test Split - <br> 80% Train <br>20% Test

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scale, 
                                                    rate, 
                                                    test_size = 0.2, 
                                                    random_state = 0)


print ("Training set has {} samples.".format(X_train.shape[0]))
print ("Testing set has {} samples.".format(X_test.shape[0]))
type(X_train)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
clf.fit(X_train, y_train) 
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred_new=[]
for e in y_pred:
    y_pred_new.append('%.4f' % e)
```

```python
len(y_pred_new)
```

```python

y_test_new = []
for e in y_test:
    y_test_new.append('%.4f' % e)
```

```python
result = pd.DataFrame({'Predictions': y_pred_new,
     'Acutual': y_test_new
    })
```
